```python
import cv2
import numpy as np
```

```python
import face_recognition
```

```python
from google.colab.patches import cv2_imshow #cv2.imshow() is disabled in Colab, because it causes Jupyter sessions to crash
```

# **Loading the IMAGE**

```python
imgElon=face_recognition.load_image_file("/content/elon1.jpg") #Importing the image
```

```python
imgElon=cv2.cvtColor(imgElon,cv2.COLOR_BGR2RGB) #converting image from BGR to RGB
```

```python
imgTest=face_recognition.load_image_file("/content/elon2.jpg") #Importing the test image
```

```python
imgTest=cv2.cvtColor(imgTest,cv2.COLOR_BGR2RGB) #converting test image from BGR to RGB
```

# **Finding the FACE**


```python
faceLoc=face_recognition.face_locations(imgElon)[0] #[0] because we are sending only one image therefore we only need its first element
encodeElon=face_recognition.face_encodings(imgElon)[0] #encoding the image
ELON=cv2.rectangle(imgElon,(faceLoc[3],faceLoc[0]),(faceLoc[1],faceLoc[2]),(255,0,255),2) #(255,0,255)->purple color and 2->thickness of the line; faceLoc[3],faceLoc[0],faceLoc[1],faceLoc[2] are the elements explained below
cv2_imshow(ELON)
#print(faceLoc) #(139, 304, 325, 118) gives 4 values top right botton left 

faceLocTest=face_recognition.face_locations(imgTest)[0] #[0] because we are sending only one image therefore we only need its first element
encodeTest=face_recognition.face_encodings(imgTest)[0] #encoding the image
ELONtest=cv2.rectangle(imgTest,(faceLocTest[3],faceLocTest[0]),(faceLocTest[1],faceLocTest[2]),(255,0,255),2) #(255,0,255)->purple color and 2->thickness of the line; faceLoc[3],faceLoc[0],faceLoc[1],faceLoc[2] are the elements explained below
cv2_imshow(ELONtest)

faceDis=face_recognition.face_distance([encodeElon],encodeTest) #face_distance compares the similarity of the two images
results=face_recognition.compare_faces([encodeElon],encodeTest)
print(results,faceDis)

R=cv2.putText(imgTest,f'{results} {round(faceDis[0],2)}',(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,0),2) #cv2.putText(img,values,startPos,font,scale,color,thickness)
cv2_imshow(R)
```
